```python
#import library yang akan digunakan
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data dalam bentuk csv
df = pd.read_csv("/content/Data_Gabungan_Kotor.csv",sep = ';', header='infer')
```

```python
round(df.select_dtypes(include = ['float64', 'int64']).corr(),2)
```

```python
round(df.select_dtypes(include = ['float64', 'int64']).corr(),2).to_csv('Matriks.csv')
```

```python
#checking correlation between numerical variables
corr = round(df.select_dtypes(include = ['float64', 'int64']).corr(),2)
plt.figure(figsize=(30, 30))
sns.set(font_scale=2)
ax = sns.heatmap(corr, 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0,
            annot=True, annot_kws={"size": 12}, square=False)
cbar = ax.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=25)
for tick in cbar.ax.yaxis.get_major_ticks():
    tick.label2.set_fontweight('bold')
```
