```python
#import data dalam format csv
import pandas as pd

data = pd.read_csv("/content/Data_Gabungan_Kotor.csv",sep = ';', header='infer')
```

```python
#tampilkan data
data.head()
```

```python
#import library yang digunakan untuk Uji VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

# definisikan variabel bebas X
X = data.iloc[:, 2:42]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["Variabel"] = X.columns
  
# Menghitung VIF untuk setiap variabel bebas
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
                        
# Tampilkan hasil perhitungan VIF
print(vif_data)
```
