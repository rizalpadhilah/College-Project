## **Data Cleansing**

```python
#import packages yang digunakan
import pandas as pd
import numpy as np
```

### Original Data

```python
#membuat sebuah dataframe
df = pd.DataFrame({
    'email_address':['dean.winchester@gmail.com', 'sam.winchester@gmail.com', 'stan.edgar@vought.com', 
                    'peter.parker@gmail.com', 'stephen.strange@hotmail.com', 'bruce.banner@hotmail.com'],
    'start_date':['2021-01-01', '2021-02-02', '2020-09-08', '2021-12-01', '2021-11-15', '2020-10-08'],
    'end_date':['January 5, 2022', 'February 6, 2022', 'August 7, 2021', 'November 15, 2022',
               'March 9, 2022', 'October 8, 2022'],
    'score':[78,87,95,70,88,80]
})
```

```python
#print dataframe
df
```

### Task

1. Create two columns called `first_name` and `last_name`, and fill these two columns with the first name and the second name, taken from the email address. Their first and last name is separated by a dot (titik). Make sure to capitalize it the right way (first letter upper case, the rest in lower case)

2. Calculate the `duration` (IN DAYS) by subtracting `end_date` and `start_date`. It's okay to have the value as '80 days' or '100 days', **you don't need to remove the 'days' string**. 

3. Create another column called `rating`. If the score is above 80 (>80), the rating is 'B'. If the score is above 90 (>90), then the rating is 'A'. If the score is below or equal to 80 (<= 80), then the rating is 'C'.

So, at the end, your dataframe should have the following column, from left to right:
- `first_name`
- `last_name`
- `email_address`
- `start_date`
- `end_date`
- `duration`
- `score`
- `rating`

```python
#memisahkan kata dari separator "." dan mengambil kata pertama sebagai 'first_name'
df['first_name'] = df['email_address'].apply(lambda x: x.split('.')[0])
#membuat kolom sementara 
df['temp'] = df['email_address'].apply(lambda x: x.split('.')[1])
#memisahkan kata dari separator "@" dan mengambil kata pertama sebagai 'last_name'
df['last_name'] = df['temp'].apply(lambda x: x.split('@')[0]) 
```

```python
#mengubah datatype ke bentuk datetime
df['start_date'] = pd.to_datetime(df['start_date'], format = '%Y-%m-%d')
df['end_date'] = pd.to_datetime(df['end_date'], format = '%B %d, %Y')
#menghitung durasi dengan rumus duration = end_date - start_date
df['duration'] = df['end_date'] - df['start_date']
```

```python
#membuat kolom rating dengan menggunakan kondisional 
df['rating'] = np.where(df['score'] > 90, "A", np.where(df['score'] > 80, "B", "C" ))
```

```python
#cek tabel
df
```

```python
#mengubah posisi tabel dataframe yang baru
new_cols = ["first_name","last_name","email_address","start_date","end_date","duration","score","rating"]
df=df[new_cols]
#print df
df
```
