```python
list_of_title = [
    'Guardians of the Galaxy',
    'Thor Ragnarok',
    'Despicable Me',
    'Inception',
    'Cloud Atlas',
    'Mission Impossible Fallout',
    'Man of Steel',
    'Avengers Endgame'
]
```

```python
data = []

for i in list_of_title:
  datum = {}
  request_string = f"http://www.omdbapi.com/?i=tt3896198&apikey=289fabe9&t={i}"
  resp = requests.get(request_string)
  resp = json.loads(resp.content)
  datum['Title'] = resp['Title']
  datum['Year of Release'] = resp['Year']
  datum['Runtime'] = resp['Runtime']
  datum['Director'] = resp['Director']
  datum['Rotten Tomatoes Rating'] = resp['Ratings'][1]['Value']
  datum['Plot'] = resp['Plot']
  data.append(datum)
  
```

```python
df = pd.DataFrame.from_dict(data)
df
```
