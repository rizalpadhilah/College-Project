## Permasalahan
Kelas Pemodelan Matematika diikuti oleh 50 mahasiswa. Kelas tersebut akan dibagi menjadi 10 kelompok. Setiap kelompok terdiri dari 5 orang. Gunakan simulasi Monte Carlo untuk menentukan peluang 2 orang tertentu (misalkan Adi dan Budi) berada pada pada kelompok yang sama.

```python
import random
import numpy as np

m = int(input("Masukan banyak pengulangan simulasi = "))
n = int(input("Masukan banyak percobaan = "))
list_peluang=[]
for j in range(m):
    jumlah = 0 #inisialisasi jumlah
    print("Simulasi ke-",j+1)
    for i in range(n):
        bilangan=random.sample(range(1,50),5)#Memunculkan bilangan acak
        print("No.",i+1, ":", bilangan)
        #Misalkan angka "1" adalah Adi dan "2" adalah Budi
        if 1 in bilangan:
            if 2 in bilangan:
                jumlah = jumlah + 1
    print("Jumlah : ", jumlah)
    print("Besar Peluang :", jumlah*10/n)
    list_peluang.append(jumlah*10/n)
list_peluangnp = np.array(list_peluang)
print("Hasil peluang tiap simulasi : ", list_peluang)
print("Rata-rata peluang : ", np.mean(list_peluangnp))
print("Peluang secara teoritis : 0.081")
```

# Kesimpulan
Dari hasil simulasi monte carlo sebanyak 50 kali diperoleh rata-rata peluang Adi dan Budi selalu bersama adalah 0.0820
